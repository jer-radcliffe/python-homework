```python
"""

Ichiben Rame - Analysis on per-product basis.

Math up sales data against menu items to determine metrics per menu-item.

Metrics include:

    - count
    - revenue
    - cogs
    - profit
    
"""
from pathlib import Path
import csv

# Set the input file path irregardless of OS
filepath = Path("resources/menu_data.csv")

#menu list
menu = []

#load csv and iterate rows to add to menu list
with open(filepath, 'r') as csv_file:
    
    #read csv data
    csvreader = csv.reader(csv_file, delimiter=',')
    
    #skip header
    header = next(csvreader)
    
    #iterate through csv
    for row in csvreader:
        menu.append(row)    


# Set the input file path irregardless of OS
filepath = Path("resources/sales_data.csv")
        
#sales list        
sales = []

#load csv for sales data to add to sales list
with open(filepath, 'r') as csv_file:
    
    #read csv data
    csvreader = csv.reader(csv_file)
    
    #skip header
    header = next(csvreader)
    
    #iterate through csv
    for row in csvreader:
        sales.append(row)

#initialize tracking variables
report = {}
quantity = 0

#build out report
for record in sales:
    
    quantity = int(record[3]) #quantity
    sales_item = record[4]  #menu item
    
    #if the item has not been added to reporting dictionary add it
    if sales_item not in report:
        #initiate metric dictionary
        metrics = {
        "01-count": 0,
        "02-revenue": 0,
        "03-cogs": 0,
        "04-profit": 0,
        }
        report[sales_item] = metrics    
    
    #loop menu items
    for row in menu:
        item = row[0]
        price = float(row[3])
        cost = float(row[-1])
        
        if sales_item == item:
            report[sales_item]["01-count"]  +=quantity
            report[sales_item]["02-revenue"]+= price * quantity
            report[sales_item]["03-cogs"]   += cost * quantity
            profit = price-cost
            report[sales_item]["04-profit"] += profit * quantity
        else:
            '''
            Following line is commented out b/c it was causing the program to take too long to render results
            Per Hui in class on 11/4/20 - she told me to leave this commented out
            '''
            #print(f"{sales_item} does not equal {item}!  NO MATCH!")
            pass
            
def export():
    print(f"Results saved to {Path.cwd()}/results.txt")
    f = open("results.txt", "w")
    for menu_item,metrics in report.items():
        f.write('{} {}\n'.format(menu_item,metrics))
    f.close()

export()
            
```
